```python
#Matplotlib cheat sheet
import matplotlib.pyplot as plt    #commonly used alias
import numpy as np


# Set a black background
plt.style.use('dark_background')

x_data=np.random.random(50)*100        #50 random floats from 0 to 100
y_data=np.random.random(50)*100

#Scatter Plot
plt.scatter(x_data,y_data, marker="*",c="#ffffff",s=100,alpha=0.7)

'''
marker- style of point on plot
alpha- opacity of marker
c- color of marker
s- size of marker
'''
plt.show()
```

```python
years=[2004 + x for x in range(20)]
weight=[3,4,6,8,10,13,15,18,23,25,26,27,30,34,38,43,44,46,47,46]

#Line Chart  (default)
plt.plot(years,weight,lw=1,c="cyan",linestyle="--")
plt.show()
'''
lw- line thickness
c and linestyle can be combined like "r--" for red dashed line
'''
```

```python
#Bar graph
plt.bar(years,weight,color="teal", edgecolor="y",width=0.5,lw=1)
'''
c cannot be used, use color instead.
width-  thickness of each bar
align- center/edge
lw- line width
'''
plt.show()
```

```python
#Histogram
x=np.random.random(100)
plt.hist(x,bins=70,color="#7f28d1")
plt.show()
'''
The towers or bars of a histogram are called bins.
bins- used to specify number of bars explicitly.
    - or show specific collection of bins in the plot.
'''
```

```python
#Piechart
langs=["Python","C","C++","Java","JavaScript"]
votes=[52,20,24,10,40]
explodes=[0,0,0,0.1,0]
plt.pie(votes, labels=langs,explode=explodes,autopct="%.1f%%",pctdistance=0.8,startangle=90)
plt.show()
'''
explode- used to slice a field from piechart.
autopct- used to show calculated percentages to specific digits after decimal.
pctdistance- used to distance the percentages from the center
startangle- specifies the starting position of chart, measuring from +ve x-axis anti-clockwise.
'''
```

```python
#Boxplot
heights=np.random.normal(165,8,300)
plt.boxplot(heights)
plt.show()
'''
- top most line indicates maximum
- line b/w top most and colored line indicates 3rd Quartile
- colored line indecates median
- line b/w bottom most and colored line indicates 1st Quartile
- bottom most line indicates minimum
- circles indicate outliers
'''
```

```python
#Subplots

x=np.arange(50)
fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sine Wave",fontsize=10)

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cosine Wave",fontsize=10)

axs[1,0].plot(x, x**6)
axs[1,0].set_title("Random Function",fontsize=10)
#Values out of range

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("Log Function",fontsize=10)

plt.tight_layout()             #no overlapping of text
plt.show()
```

```python
plt.style.use('dark_background')  
X_k_list = range(1, 170, 10)  
Y_p_list = [ float(x)/170.0 for x in range(1, 170, 10) ]
fig = plt.figure(figsize=plt.figaspect(0.5))  
ax = fig.add_subplot(1, 1, 1, projection='3d')
X, Y = np.meshgrid(X_k_list, Y_p_list)  
def critical_function(b, c):  
    num = random.uniform(0, 1) * 10.0  
    return num + (b * c)   
  
Z_accuracy = X.copy()  
Z_accuracy = Z_accuracy.astype(np.float32)  
for i in range(len(X_k_list)):  
    for j in range(len(Y_p_list)):  
        Z_accuracy[j][i] = critical_function(Y_p_list[j], X_k_list[i])  
  
surf = ax.plot_surface(X, Y, Z_accuracy,   
    rstride=1, cstride=1, cmap="viridis",  
    linewidth=0, antialiased=False)  
  
ax.set_xlabel('X')  
ax.set_ylabel('Y')  
ax.set_zlabel('Z')  
fig.colorbar(surf, shrink=0.5, aspect=10)  
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Set a black background
plt.style.use('dark_background')

# Create data for the four surface plots
X1, Y1 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z1 = np.sin(np.sqrt(X1**2 + Y1**2))

X2, Y2 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z2 = np.cos(np.sqrt(X2**2 + Y2**2))

X3, Y3 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z3 = np.exp(-(X3**2 + Y3**2))

X4, Y4 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z4 = np.sin(X4) * np.cos(Y4)

# Create subplots for the four surface plots
fig = plt.figure(figsize=(12, 9))

# Plot 1
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
surf1 = ax1.plot_surface(X1, Y1, Z1, cmap='viridis')
ax1.set_title('Surface Plot 1')

# Plot 2
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
surf2 = ax2.plot_surface(X2, Y2, Z2, cmap='plasma')
ax2.set_title('Surface Plot 2')

# Plot 3
ax3 = fig.add_subplot(2, 2, 3, projection='3d')
surf3 = ax3.plot_surface(X3, Y3, Z3, cmap='magma')
ax3.set_title('Surface Plot 3')

# Plot 4
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
surf4 = ax4.plot_surface(X4, Y4, Z4, cmap='inferno')
ax4.set_title('Surface Plot 4')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
